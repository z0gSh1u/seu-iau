```python
import numpy as np
import matplotlib.pyplot as plt
from helper import Gabor
from PIL import Image
import cv2
```

```python
img = Image.open('./materials/tvprogram.png').convert('L')
img = np.array(img)
plt.figure()
plt.imshow(img, cmap='gray')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
for i, theta in enumerate([0, 45, 90, 135]):
    g = Gabor(3, np.radians(theta), 5, 0, 1)
    imgFiltered = cv2.filter2D(img, None, g)
    plt.subplot(2, 2, i + 1)
    plt.imshow(imgFiltered, cmap='gray'); plt.title(r'$\theta=${} Deg'.format(theta))
plt.show()
```

```python
img = np.array(Image.open('./materials/fingerprint.jpg').convert('L'), dtype=np.uint8).astype(np.float32)
plt.figure()
plt.imshow(img, cmap='gray')
plt.show()
```
